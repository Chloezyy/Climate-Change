```python
deco = sm.tsa.seasonal_decompose(y, model='additive')
fig = deco.plot()
fig.set_size_inches(18.5, 10.5)
plt.show()
```

```python
#residual看起来平稳杂乱-合理
deco.resid.describe()
#确定mean=0
```

```python
deco.resid.isnull().sum()
```

```python
CO2Residual = deco.resid
CO2Residual = CO2Residual.fillna(CO2Residual.bfill())
CO2Residual = CO2Residual.fillna(CO2Residual.ffill())
```

```python
#residual也当成时间序列的数据， AR,MA,ARMA都可以对平稳数据分析
CO2Residual.describe()
```

```python
fig = CO2Residual.plot()
plt.show() 
```

```python
#检验co2 residue 是否是平稳的
adf_co2res = adfuller(CO2Residual)
print(adf_co2res)
```

```python
print('ADF Statistic: %f' % adf_co2res[0])
print('p-value: %f' % adf_co2res[1])
#假设不平稳的话，出现这个值的概率有多大 -- data不平稳不对--data平稳

print('Critical Values:')
for key, value in adf_co2res[4].items():
	print('\t%s: %.3f' % (key, value))
#alpha不同的z值

#数据平稳，所以ts-ARMA可以上
```

```python
##回到原始数据！

from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

#plot_acf(CO2Residual,title="CO2")  
#residu是平稳的，对平稳数据不再做分析了（res算完AR再加回去trend+season），太麻烦。除非以后做refine/research

plot_acf(y,lags=75,title="CO2")
#tail off托尾明显，如果很长- 做differential差分
```

```python
#plot_pacf(CO2Residual,title="CO2")
plot_pacf(y,lags=75,title="CO2")
#拖，但有明显下降趋势

#所以还是要脱差分，数据绝对不平稳
```

```python
mod = sm.tsa.statespace.SARIMAX(y,
                                order=(1, 1, 1),
				)
#dpq =111-改!!!!!!，跑得慢 ，seasonal order可以写12（一年）

TSresults = mod.fit()

#print(TSresults.summary().tables[1])

TSresults.plot_diagnostics(figsize=(15, 12))
plt.show()
```

```python
#模型train完了得到参数们-phi， theta


##########数据只取到2001，希望我取到2023，再看2019往后预测的如何
#################


#预测，假设现在1998，预测的数据有多大差
pred = TSresults.get_prediction(start=pd.to_datetime('1998-01-01'), dynamic=False)
#  dynamic=False)== 用前置的真实值

pred_ci = pred.conf_int()
#Returns the confidence interval of the fitted parameters.

ax = y['1990':].plot(label='observed')
pred.predicted_mean.plot(ax=ax, label='Forecast', alpha=.6)

ax.fill_between(pred_ci.index, pred_ci.iloc[:, 0], pred_ci.iloc[:, 1], color='k', alpha=.2)
ax.set_xlabel('Date')
ax.set_ylabel('CO2 Levels')
plt.legend()

plt.show()

#灰色是预测的区间值，桔红色是预测的中值
```

```python
pred_ci
```

```python
y_forecasted = pred.predicted_mean
y_truth = y['1998-01-01':]

# Compute the mean square error
mse = ((y_forecasted - y_truth) ** 2).mean()
print('The Mean Squared Error of our forecasts is {}'.format(round(mse, 2)))

#衡量好不好 - 
```

```python
#pred_dynamic = TSresults.get_prediction(start=pd.to_datetime('1998-01-01'), dynamic=False, full_results=True)
pred_dynamic = TSresults.get_prediction(start=pd.to_datetime('1998-01-01'), dynamic=True, full_results=True)
pred_dynamic_ci = pred_dynamic.conf_int()

# dynamic=True 预测的值-预测下一个值，高的值预测更高，低的值预测更低，往往更加发散
#但也算有一定的指导意义（最低最高）


ax = y['1990':].plot(label='observed', figsize=(10, 8))
pred_dynamic.predicted_mean.plot(label='Dynamic Forecast', ax=ax)
ax.fill_between(pred_dynamic_ci.index,
                pred_dynamic_ci.iloc[:, 0],
                pred_dynamic_ci.iloc[:, 1], color='r', alpha=.25)

ax.fill_betweenx(ax.get_ylim(), pd.to_datetime('1998-01-01'), y.index[-1],
                 alpha=.1, zorder=-1)

ax.set_xlabel('Date')
ax.set_ylabel('CO2 Levels')

plt.legend()
plt.show()
```
